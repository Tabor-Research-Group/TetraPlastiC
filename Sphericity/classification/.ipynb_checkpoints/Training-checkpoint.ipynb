```python
import pandas as pd
from sklearn import svm
import numpy as np
import re
import matplotlib.pyplot as plt
import pickle
```

```python
def string_to_pandas_series_ops(string,variable):
    for variable_name in variable.keys():
        variable_ref = f"{variable_name} = variable['{variable_name}']"
        exec(variable_ref)
    # Create a regular expression pattern to match variable names
    pattern = r"[a-zA-Z]+\d+"

    # Replace the original string with a modified string that uses the variable references
    modified_string = re.sub(pattern, lambda match: f"{match.group()} ", string)

    # Evaluate the modified string using the `eval` function
    result = eval(modified_string)

    # Return the result and the variable references
    return result
```

```python
j = 0
df_train = pd.read_csv(str(j)+'/train.dat',sep='\t')
df_test = pd.read_csv(str(j)+'/test.dat',sep='\t')
df_test2 = pd.read_csv(str(j)+'/test2.dat',sep='\t')
Class_test = []
for i in range(len(df_test.iloc[:,0])):
    Class_test.append(int(df_test.iloc[:,0][i].split('group')[-1]))
Class = []
for i in range(len(df_train.iloc[:,0])):
    Class.append(int(df_train.iloc[:,0][i].split('group')[-1]))

variable_names = ["feature"+str(i) for i in range(1,df_train.shape[1],1)]
variable_ref = {variable_name: df_train[variable_name] for variable_name in variable_names}
variable_test = {variable_name: df_test[variable_name] for variable_name in variable_names}
variable_test2 = {variable_name: df_test2[variable_name] for variable_name in variable_names}


model_file = open('../train_'+str(j)+'/models/top0100_D002','r')
feature_space = open('../train_'+str(j)+'/SIS_subspaces/Uspace.expressions','r')
lines_model = model_file.readlines()
lines_feature = feature_space.readlines()
k = 0
ID1 = int(lines_model[k+1].split()[4])
ID2 = int(lines_model[k+1].split()[5].replace(')',''))
```

```python
Y_pred_score = []
Selected_features = []
for i in range(100):
    df_train = pd.read_csv(str(i)+'/train.dat',sep='\t')
    df_test = pd.read_csv(str(i)+'/test.dat',sep='\t')
    df_test2 = pd.read_csv(str(i)+'/test2.dat',sep='\t')
    Class_test = []
    for j in range(len(df_test.iloc[:,0])):
        Class_test.append(int(df_test.iloc[:,0][j].split('group')[-1]))
    Class = []
    for j in range(len(df_train.iloc[:,0])):
        Class.append(int(df_train.iloc[:,0][j].split('group')[-1]))

    variable_names = ["feature"+str(l) for l in range(1,df_train.shape[1],1)]
    variable_ref = {variable_name: df_train[variable_name] for variable_name in variable_names}
    variable_test = {variable_name: df_test[variable_name] for variable_name in variable_names}
#     variable_test2 = {variable_name: df_test2[variable_name] for variable_name in variable_names}


    model_file = open(str(i)+'/models/top0100_D002','r')
    feature_space = open(str(i)+'/SIS_subspaces/Uspace.expressions','r')
    lines_model = model_file.readlines()
    lines_feature = feature_space.readlines()
    
    with open('./feature_name_'+str(i), "rb") as fp:
        feature_name = pickle.load(fp)
    for j in range(100):
        ID1 = int(lines_model[j+1].split()[4])
        ID2 = int(lines_model[j+1].split()[5].replace(')',''))

        Feature1_string = lines_feature[ID1-1].split()[0]
        Feature2_string = lines_feature[ID2-1].split()[0]
        feature1 = string_to_pandas_series_ops(Feature1_string, variable_ref)
        feature2 = string_to_pandas_series_ops(Feature2_string, variable_ref)

        numbers = re.findall(r'feature(\d+)', Feature1_string)
        numbers = list(map(int, numbers))
        score_number = len(numbers)
        for k in numbers:
            Selected_features.append((feature_name[k-1]))
            
        numbers = re.findall(r'feature(\d+)', Feature2_string)
        numbers = list(map(int, numbers))
        score_number += len(numbers)
        for k in numbers:
            Selected_features.append((feature_name[k-1]))

        feature1_test = string_to_pandas_series_ops(Feature1_string, variable_test)
        feature2_test = string_to_pandas_series_ops(Feature2_string, variable_test)

#         feature1_test2 = string_to_pandas_series_ops(Feature1_string, variable_test2)
#         feature2_test2 = string_to_pandas_series_ops(Feature2_string, variable_test2)

        X = np.array([feature1, feature2])
        X = X.transpose()

        X_test = np.array([feature1_test,feature2_test])
        X_test = X_test.transpose()

#         X_test2 = np.array([feature1_test2,feature2_test2])
#         X_test2 = X_test2.transpose()

        Y = np.array(Class)
        Y_test = np.array(Class_test)

        best_C = 0
        best_G = 0
        best_score = 0

        for c in [0.1,1,10,100,1000]:
            for g in [0.001,0.01,0.1,1,10]:
                svf_rbf = svm.SVC(C = c, gamma = g,probability=True)
                svf_rbf.fit(X, Y)
                if svf_rbf.score(X,Y) > best_score:
                    best_C = c
                    best_G = g

        svf_rbf = svm.SVC(C = best_C, gamma = best_G,probability=True)
        svf_rbf.fit(X, Y)
        for i in range(score_number):
            Y_pred_score.append(svf_rbf.score(X_test,Y_test))
#     Y_pred2 = svf_rbf.predict(X_test2)
#     Y_pred2_all.append(Y_pred2)
```

```python
selected_feature, feature_counts = np.unique(Selected_features, return_counts=True)
```

```python
score_stacking = []
for i in selected_feature:
    tmp = np.zeros(6)
    tmp_indices = np.where(np.array(Selected_features) == i)[0]
    all_scores = np.array(Y_pred_score)[tmp_indices]
    for j in all_scores:
        if j >=0.8:
            tmp[5] += 1
        elif j >= 0.6 and j < 0.8:
            tmp[4] += 1
        elif j >= 0.4 and j < 0.6:
            tmp[3] += 1
        elif j >= 0.2 and j < 0.4:
            tmp[2] += 1
        elif j >= 0 and j < 0.2:
            tmp[1] += 1
        else:
            tmp[0] += 1
    score_stacking.append(tmp)
```

```python
fig, ax = plt.subplots()
labels = ['score < 0.0', '0.2 <= score < 0.0', '0.4 <= score < 0.2','0.6 <= score < 0.4',
                                   '0.8 <= score < 0.6','1.0 <= score < 0.8']
width = 0.5
for i in range(6):
    bottom = np.sum(np.array(score_stacking)[:,:i],axis=1)/160000
    ax.bar(selected_feature, np.array(score_stacking)[:,i]/160000,width, label=labels[i], bottom=bottom)

plt.xticks(rotation='vertical')
plt.xlabel('Features')
plt.ylabel('Ratio in the SISSO Model Ensemble')
ax.legend()
plt.title('Selected Feature Ratio from Classification')
```
