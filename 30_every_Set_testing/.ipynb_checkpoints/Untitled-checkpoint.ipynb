```python
import pandas as pd
from urllib.request import urlopen
from urllib.parse import quote
import os
import numpy as np
import pickle 
from sklearn.preprocessing import MinMaxScaler
```

```python

df_PC = pd.read_csv('../PC_v9.csv', encoding='latin-1')
PC_GCM = df_PC.iloc[:,9:37]
PC_SYM = pd.concat([df_PC.iloc[:,39], df_PC.iloc[:,41:43],df_PC.iloc[:,44:46],df_PC.iloc[:,47]],axis=1)
PC_H = df_PC.iloc[:,49:57]
PC_R = df_PC.iloc[:,58:61]
PC_Str = df_PC.iloc[:,72]
PC_Htr = df_PC.iloc[:,73]
PC_Ttr = df_PC.iloc[:,74]
PC_Tm = df_PC.iloc[:,78]
PC_year = df_PC.iloc[:,80]

PC_SVI = df_PC.iloc[:,62:65]
PC_EDA = df_PC.iloc[:,65:70]
```

```python
sample_data = pd.read_csv('30_GCM_scaled.csv')
```

```python
Properties = np.array(sample_data['Properties'])
PC_Htr = PC_Htr.dropna()
```

```python
sample_indices = []
for i in range(len(Properties)):
    tmp_idx = np.where(np.array(PC_Htr) == Properties[i])[0][0]
    sample_indices.append(tmp_idx)
```

```python
H_sample = pd.concat([PC_Htr.iloc[sample_indices],PC_H.iloc[sample_indices,:]],axis=1)
```

```python
sample_name_train = []
s1 = 1
for i in range(H_sample.shape[0]):
    sample_name_train.append('sample'+str(s1))
    s1 += 1
    
feature_name = []
feature_name.append('Properties')
for i in range(H_sample.shape[1]-1):
    feature_name.append('feature' + str(i+1) )
    
```

```python
scaler = MinMaxScaler()
H_sample_scaled = scaler.fit_transform(H_sample.iloc[:,1:])
H_sample_scaled = pd.DataFrame(H_sample_scaled)

H_sample_scaled.index = sample_name_train
H_sample.index = sample_name_train

H_sample_scaled = pd.concat([H_sample.iloc[:,0],H_sample_scaled],axis=1)


H_sample_scaled.index = sample_name_train
H_sample_scaled.columns = feature_name
H_sample_scaled.columns.name = 'Materials'
```

```python
H_sample_scaled.to_csv('30_H_scaled.csv')
```
