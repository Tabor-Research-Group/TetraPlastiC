```python
import pandas as pd
from sklearn import svm
import numpy as np
import re
import random
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
plt.rcParams['figure.dpi'] = 300
```

```python
def string_to_pandas_series_ops(string,variable):
    for variable_name in variable.keys():
        variable_ref = f"{variable_name} = variable['{variable_name}']"
        exec(variable_ref)
    # Create a regular expression pattern to match variable names
    pattern = r"[a-zA-Z]+\d+"

    # Replace the original string with a modified string that uses the variable references
    modified_string = re.sub(pattern, lambda match: f"{match.group()} ", string)

    # Evaluate the modified string using the `eval` function
    result = eval(modified_string)

    # Return the result and the variable references
    return result
```

```python
Regression_year = ['1969.0', '1970.0', '1973.0', '1975.0', '1976.0' ,
          '1982.0', '1985.5', '1986.0', '1988.0', '1991.5', '1994.0', '1994.6666670000002', '1995.0',
          '1995.5', '1996.0', '1997.5', '1999.0', '2001.0', '2005.0', '2006.0', '2007.5',
          '2014.0', '2017.0' ]
```

```python
Train_year = []
all_test_loss = []
for i in Regression_year:
    df_train = pd.read_csv('S/'+i+'/train.dat',sep='\t')
    df_test = pd.read_csv('S/'+i+'/test.dat',sep='\t')
    H_train = df_train['Properties']
    H_test = df_test['Properties']
    variable_names = ["feature"+str(i) for i in range(1,df_train.shape[1]-1,1)]
    variable_ref = {variable_name: df_train[variable_name] for variable_name in variable_names}
    variable_test = {variable_name: df_test[variable_name] for variable_name in variable_names}


    model_file = open('S/'+i+'/models/top0100_D002','r')
    feature_space = open('S/'+i+'/SIS_subspaces/Uspace.expressions','r')
    lines_model = model_file.readlines()
    lines_feature = feature_space.readlines()
    model_test_loss = []
    
    for j in range(100):
        ID1 = int(lines_model[j+1].split()[4])
        ID2 = int(lines_model[j+1].split()[5].replace(')',''))

        Feature1_string = lines_feature[ID1-1].split()[0]
        Feature2_string = lines_feature[ID2-1].split()[0]
        feature1 = string_to_pandas_series_ops(Feature1_string, variable_ref)
        feature2 = string_to_pandas_series_ops(Feature2_string, variable_ref)

        feature1_test = string_to_pandas_series_ops(Feature1_string, variable_test)
        feature2_test = string_to_pandas_series_ops(Feature2_string, variable_test)

        X = np.array([feature1, feature2])
        X = X.transpose()

        X_test = np.array([feature1_test,feature2_test])
        X_test = X_test.transpose()

        Y = np.array(H_train)
        Y_test = np.array(H_test)
        random.seed(0)
        train_idx = []
        val_idx = []
        train_idx.append(random.sample(range(X.shape[0]), int(X.shape[0]*0.8)))
        train_idx = [y for x in train_idx for y in x]
        val_idx = [x for x in range(X.shape[0]) if x not in train_idx]
        train_idx.sort()
        val_idx.sort()

        best_e = 0.001
        best_c = 0.001
        best_g = 0.001
        best_score = 0
        for e in [0.001, 0.1,1,10,100,1000]:
            for c in [0.001,0.1,1,10,100,1000]:
                for g in [0.001,0.01,0.1,1,10,100]:
                    svf_rbf = svm.SVR(C = c, gamma = g, epsilon = e)
                    svf_rbf.fit(X[train_idx], Y[train_idx])
                    score = svf_rbf.score(X[val_idx],Y[val_idx])
                    if  score > best_score:
                        best_e = e
                        best_c = c
                        best_g = g
                        best_score = score
        svf_rbf = svm.SVR(C = best_c, gamma = best_g,epsilon = best_e)
        svf_rbf.fit(X, Y)
        Y_pred = svf_rbf.predict(X_test)
        model_test_loss.append(mean_squared_error(Y_test,Y_pred,squared=False))
    all_test_loss.append(np.mean(model_test_loss))
    Train_year.append(float(i))
```

```python
Train_size = []
for i in Regression_year:
    df_train = pd.read_csv('S/'+i+'/train.dat',sep='\t')
    Train_size.append(df_train.shape[0])
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.scatter(Train_year, all_test_loss, c = '#2a9d8f')
ax1.plot(Train_year, all_test_loss, c = '#2a9d8f')
ax2.scatter(Train_year, Train_size, c='#f4a261')
ax2.plot(Train_year, Train_size, c='#f4a261')

ax1.set_xlabel('Trained Year',fontsize=18)
ax1.set_ylabel('Loss(RMSE) (cal/mol/K)', color='#2a9d8f',fontsize=18)
ax1.tick_params(axis='x', labelsize=14)
ax1.tick_params(axis='y', labelsize=14)
ax2.set_ylabel('Data Size', color='#f4a261',fontsize=18)
ax2.tick_params(axis='y', labelsize=14)
plt.title('Entropy Prediction',fontsize=18)
```
